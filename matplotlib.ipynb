```
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 5)
y = x**3
plt.subplot(1,2,1)
a = [1,2,3,4]
b = [1,4,9,16]
plt.plot(a, b, 'go')

plt.subplot(1,2,2)
plt.plot(x, y, 'r^')

plt.title('2nd subplot')
plt.suptitle('My sub-plots')

plt.show()
```

```

plt.subplot(2,1,1)
plt.plot(a, b, 'go')
plt.title('1st subplot')

plt.subplot(2, 1, 2)
plt.plot(a, b, 'r^')
plt.title('2nd subplot')

plt.suptitle('My subplots')
plt.show()
```

```
x = np.arange(0, 10)
y = x ** 3

x2 = np.arange(0, 10)
y2 = x2 ** 2

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(16, 16))

divisions = ['Div-A', 'Div-B', 'Div-C', 'Div-D', 'Div-E']
division_average_marks = [70, 82, 73, 65, 68]
axes[1,1].bar(divisions, division_average_marks)
axes[1,1].set_xlabel('marks')
axes[1,1].set_ylabel('division')

axes[0,1].plot(x, y, 'go')
axes[1,0].plot(x2, y2, 'r^')
axes[0,1].set_title('Cubes')
axes[1,0].set_title('Squares')
```

```
divisions = ['Div-A', 'Div-B', 'Div-C', 'Div-D', 'Div-E']
division_average_marks = [70, 82, 73, 65, 68]
variance = [35, 2, 7, 6, 4]
plt.barh(divisions, division_average_marks, xerr=variance, color='green')
plt.title('bar graph')
plt.xlabel('Marks')
plt.ylabel('Divisions')
```

```
divisions = ['Div-A', 'Div-B', 'Div-C', 'Div-D', 'Div-E']
division_average_marks = [70, 82, 73, 65, 68]
boys_average_marks = [68, 67, 77, 61, 70]
index = np.arange(5)
width = 0.30

plt.bar(index-width/2, division_average_marks, width, color='green', label='Division Marks')
plt.bar(index+width/2, boys_average_marks, width, color='blue', label='Boys Marks')

plt.title('bar graph')
plt.xlabel('Marks')
plt.ylabel('Divisions')

plt.legend()
plt.show()
```

```
divisions = ['Div-A', 'Div-B', 'Div-C', 'Div-D', 'Div-E']
boys_average_marks = np.array([68, 67, 77, 61, 70])
girls_average_marks = np.array([72, 92, 69, 65, 68])

index = divisions
width = 0.30

plt.bar(index, boys_average_marks, width, color='blue', label='Boys Marks')
plt.bar(index, girls_average_marks, width, color='red', label='Girls Marks', bottom=boys_average_marks)

values = boys_average_marks + girls_average_marks
for pos, value in zip(index, boys_average_marks + girls_average_marks):
  plt.text(pos, value + 1, str(value), fontweight='bold', horizontalalignment='center')
```

```
import string
firms = ['Firm ' + i for i in string.ascii_uppercase[:5]]
market_share = (np.random.random(5) * 20 + 10).astype(int)
market_share = [20,25,15,10,20]
print(market_share)
explode = [0, 0.1, 0, 0, 0]
plt.pie(market_share, explode=explode, labels=firms, shadow=True, startangle=45)
plt.axis('equal')
```

```
import numpy as np
x = np.random.randn(10000)
plt.title('Histogram title')
plt.xlabel('data value')
plt.ylabel('data frequency')
plt.hist(x, bins=50)
plt.show()
```

```
height = (np.array(np.random.random(15) * 60 + 140)).astype(int)
weight = (np.array(np.random.random(15) * 40 + 60)).astype(int)
plt.xlim(140, 200)
plt.ylim(60, 100)
plt.scatter(height, weight)
plt.title('Scatter Plot')
plt.xlabel('Height')
plt.ylabel('Weight')

plt.show()
```

```
from mpl_toolkits import mplot3d
height = (np.array(np.random.random(15) * 60 + 140)).astype(int)
weight = (np.array(np.random.random(15) * 40 + 60)).astype(int)
ax = plt.axes(projection='3d')
ax.scatter3D(height, weight)
ax.set_xlabel('Height')
ax.set_ylabel('Weight')
plt.show()
```
